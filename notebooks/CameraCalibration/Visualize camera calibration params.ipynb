```python
DATA_DIR = Path('/mnt/sda1/SMLAT/data/real_data/dark_images/camera_calibration_pool/')
```

### Let's calibrate one pixel

```python
pixel_no = [632, 436]
exposures = [25, 75, 150, 300]
values = {}
for exp in exposures:
    data_dir = DATA_DIR / Path('raw_' + str(exp) + 'ms')
    values[exp] = []
    for i in range(5000):
        img_path = data_dir / Path('img_' + str(i).zfill(9) + '.tif')
        img = imread(img_path)
        values[exp].append(img[pixel_no[0], pixel_no[1]])
        print(img_path)
    print(f"Exp: {exp} done")
```

```python
means = []
variances = []
for exp in values:
    means.append(np.mean(values[exp]))
    variances.append(np.var(values[exp]))
```

```python
variances
```

```python
plt.figure()
plt.plot(exposures, means)
plt.show()
```

```python
plt.figure()
plt.plot(exposures, variances)
plt.show()
```
